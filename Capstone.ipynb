```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv("adult.csv",na_values=' ?')
a
```

```python
a=a.loc[1:,:]
```

```python
col_name=a.columns.values
for i in col_name:
    a[i].replace(to_replace=np.nan,value=a[i].mode()[0],inplace=True)
a
```

```python
a.plot(kind="box",figsize=(10,7))
```

```python
a["occupation"].groupby(by=a["education"]).count()
```

```python
b=a.drop(columns=['capital-gain','capital-loss'])
c=['age','fnlwgt','education-num','hours-per-week']
for i in col_name:
    q1=b.quantile(0.25)
    q3=b.quantile(0.75)
    iqr=q3-q1
    max_val=q3+iqr*1.5
    #print(iqr*1.5)
    min_val=q1-iqr*1.5
#print(min_val)
for i,j,k in zip(min_val,max_val,c):
    b=b[b[k]>=i]
    b=b[b[k]<=j]
b.shape
```

```python
b.describe()
```

```python
#b.plot.line(x='workclass',y=['age','fnlwgt'])
#b.plot.line()
b['kd']=pd.cut(b['hours-per-week'],bins=[30,40,50,60],labels=[40,50,60])
k=b.groupby(['salary','kd']).size()
k.plot(kind="bar")
```

```python
b['age_grp']=pd.cut(b['age'],bins=[10,18,28,50,78],labels=['teen','bachelor','adult','old'])
k1=b.groupby(['age_grp','salary']).size()
k1.plot(kind='bar')
```

```python
b.corr()
```
