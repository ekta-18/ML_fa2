```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
kl_rahul = np.random.randint(112, size = 10)
kl_rahul
```

```python
hit_man = np.random.randint(116, size = 10)
hit_man
```

```python
matches = np.arange(1,11)
matches
```

```python
kl_mean = np.mean(kl_rahul)
kl_mean
```

```python
hit_mean = np.mean(hit_man)
hit_mean
```

```python
hit_mean_lst = np.full(10, hit_mean)
kl_mean_lst = np.full(10, kl_mean)
```

```python
plt.plot(matches, kl_rahul, label = 'KL_Rahul')
plt.plot(matches, hit_man, label = 'Hitman')
plt.plot(matches, kl_mean_lst, linestyle = 'dotted', label = 'KL_Rahul_mean')
plt.plot(matches, hit_mean_lst,  linestyle = 'dotted', label = 'Hitman_mean')
plt.xlabel("Matches")
plt.ylabel("Runs")
plt.legend()
plt.show()
```
